2. Creating a Timestamped Backup Automatically

```python
# 2. Creating a Timestamped Backup Automatically
from datetime import datetime
import shutil
from pathlib import Path

def create_backup(original_path):
    original_path = Path(original_path) # Ensure it's a Path object
    
    if original_path.exists():
        # Create a unique filename using the current date and time
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        backup_name = f"{original_path.stem}_backup_{timestamp}{original_path.suffix}"
        backup_path = original_path.parent / backup_name
        
        # Copy the original file to the backup location
        shutil.copy(original_path, backup_path)
        print(f"Backup created: {backup_path}")
    else:
        print("No original file found to back up.")

# Test the function with your activity file
# Replace '2025-0495' with your actual Student ID if different
student_id = "2025-0495"
test_file = Path.home() / "Documents" / "Lunjas_Activity_5" / f"intro_{student_id}.txt"
create_backup(test_file)
```

3. Handling User Input Errors

```python
# 3. Handling User Input Errors
# When taking input for filenames or options, it's important to validate.

filename = input("Enter filename: ").strip()

if not filename.endswith(".txt"):
    print("Warning: Filename should end with '.txt'. Appending extension.")
    filename += ".txt"

print(f"Final validated filename: {filename}")
```

4. Detecting File Type and Size Before Reading

```python
# 4. Detecting File Type and Size Before Reading (FIXED)
from pathlib import Path

# Specify the file you want to check
file_path = Path("data.txt")

# 1. First, check if the file actually exists to avoid a crash
if file_path.exists():
    # 2. Now it is safe to check the extension and size
    if file_path.suffix == ".txt" and file_path.stat().st_size < 100000:
        with open(file_path, "r", encoding="utf-8") as f:
            print(f.read())
    else:
        print("File is either too large or not a .txt file.")
else:
    # 3. Handle the case where the file is missing
    print(f"Error: The file '{file_path}' was not found in the directory.")
```

5. Saving Logs to a File

```python
# 5. Saving Logs to a File
# Track file operations for review
from datetime import datetime

def log_action(action, filename):
    # The "a" mode is used here to ensure we append to the log 
    # rather than overwriting previous entries.
    with open("file_log.txt", "a", encoding="utf-8") as log:
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        log.write(f"[{timestamp}] {action} - {filename}\n")

# Test the logging function
log_action("Created Backup", "intro_2025-0357.txt")
log_action("Validated Filename", "Lab5_Final_Report.txt")

print("Actions have been logged to file_log.txt")
```
